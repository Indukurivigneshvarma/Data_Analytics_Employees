```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Bluepal\Desktop\Data_Analytics_Employees\Excell_sheets\clean_employee_dataset.csv", sep=';')

def safe_split_multi_sep(x):
    if pd.isna(x):
        return []
    x = str(x).replace(';', ',')      
    return [i.strip().title() for i in x.split(',') if i.strip() != '']

df['department'] = df['department'].apply(safe_split_multi_sep)
df['skills'] = df['skills'].apply(safe_split_multi_sep)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 1, figsize=(12, 12), gridspec_kw={'height_ratios':[2, 1]})

sns.kdeplot(df['age'], fill=True, color='skyblue', linewidth=2, alpha=0.6, ax=axes[0])
axes[0].set_title('Age Distribution of Employees (KDE)', fontsize=16, pad=20)
axes[0].set_xlabel('')
axes[0].set_ylabel('Density', fontsize=14)

sns.boxplot(x='gender', y='age', data=df, palette='pastel', ax=axes[1])
axes[1].set_title('Age Distribution by Gender', fontsize=16, pad=20)
axes[1].set_xlabel('Gender', fontsize=14)
axes[1].set_ylabel('Age', fontsize=14)
plt.tight_layout(h_pad=3)  
```
